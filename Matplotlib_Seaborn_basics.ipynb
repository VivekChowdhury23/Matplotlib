# Visualization using Matplotlib

##### Matplotlib is a cross-platform, data visualization and graphical plotting library for Python and its numerical extension NumPy. As such, it offers a viable open source alternative to MATLAB. Developers can also use matplotlib's APIs (Application Programming Interfaces) to embed plots in GUI applications.

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

## Basics of Matplotlib

### Setting a style

```python
plt.style.use('classic') #https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
```

### Plotting a simple sine and cosine curve

```python
# Creating a Numpy array
x = np.linspace(0,10,100)

# Creating the Sine curve
plt.plot(x, np.sin(x))
plt.show()
```

```python
# Creating the cosine curve
plt.plot(x, np.cos(x))
plt.show()
```

### Using Subplots

#### The pseudo code for subplot is **`plt.subplot(rows,columns,panel number)`**

```python
# Plotting the a sine and a cosine graph side by side using subplots
x = np.linspace(1,10,30)
plt.figure(figsize=[20,5]) # we can use figsize and pass [breadth, height] to control the output

# For the first plot
plt.subplot(1,2,1)
plt.plot(x, np.sin(x))

# For the second plot
plt.subplot(1,2,2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
# Plotting the a sine and a cosine graph one above another using subplots
x = np.linspace(1,10,30)
plt.figure(figsize=[5,5]) # we can use figsize and pass [breadth, height] to control the output

# For the first plot
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

# For the second plot
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

plt.show()
```

### Adjusting the plot: Line Colors and styles

#### For Line Colors...

```python
# Plotting the graphs

plt.plot(x, np.sin(x), color='blue') #specify the color
plt.plot(x, np.sin(x - 1), color='g') #short color code
plt.plot(x, np.sin(x - 2), color='0.75') #greyscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') #Hex code RRGGBB from 00 to FF
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) #RGB tuple
plt.plot(x, np.sin(x - 5), color='chartreuse') #HTML color name
plt.show()
```

#### For Line Styles...

```python
# Plotting the graphs

plt.plot(x, x, linestyle='solid') #you can also use '-'
plt.plot(x, x + 1, linestyle='dashed') #you can also use '--'
plt.plot(x, x + 2, linestyle='dashdot') #you can also use '-.'
plt.plot(x, x + 3, linestyle='dotted') #you can also use '.'
plt.show()
```

### Adjusting the plot: Axes limits

#### We can limit the axes by using **`plt.xlim()`** and **`plt.ylim()`** methods

```python
# Creating a Numpy Array
x = np.linspace(0,20,60)

# Creating a sine curve
plt.figure(figsize=[10,5])

#First plot without applying limits
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.axis('equal')

#2nd plot with limits applied
plt.subplot(2,1,2)
plt.plot(x, np.sin(x))
# Now limiting the x axis: plt.xlim(start pos, end pos)
plt.xlim(-1,15)
plt.show()
```

### Labelling the plots

```python
plt.style.use('ggplot') #Changing the style

# Creating a simple sine and cosine curve
x = np.linspace(0,10,100)

plt.figure(figsize=[15,4])

# Sine curve
plt.subplot(1,2,1)
plt.plot(x, np.sin(x))
plt.title('Sine Curve', pad=20) #Adding a title to the graph
plt.xlabel('Time', color='blue') #Adding x label
plt.ylabel('Intensity', color='green') #Adding y label

# Cosine curve
plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
plt.title('Cosine Curve', pad=20) #Adding a title to the graph
plt.xlabel('Time', color='blue') #Adding x label
plt.ylabel('Intensity', color='green') #Adding y label

plt.show()
```

### Scatter Plots (to compare numerical data)

- a graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

- It can be used to identify a relationship or pattern between two quantitative variables and the presence of outliers within them.

```python
plt.style.use('seaborn-whitegrid') #Changing the style
```

```python
# Creating a numpy array and a sine graph
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y,'o',color='black')
plt.show()
```

##### This is an example of a scatter plot, let's see another example of scatter plot.

```python
# Creating a scatter plot
rng = np.random.RandomState(0)

for marker in ['o','.',',','x','+','v','^','<','>','s','d']:
    plt.plot(rng.random(5), rng.random(5), marker, label="Marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0,1.8);
```

##### Let's look at another example of scatter plot

```python
# Defining some values for x and y
x = np.arange(0,10)
y = np.arange(20,30)

# plotting the scatter plot
plt.plot(x,y, '-ok') #joining a line through the points, '-' = draw line, 'o' =circle marker, 'k' = black
plt.show()
```

#### Scatter plots using **`plt.scatter()`**

```python
# Creating x and y values
X = []
Y = []

np.random.seed(0) #seed for reproducibility

for i in range(0,11):
    x = np.random.randint(0,10)
    y = np.random.randint(20,40)
    X.append(x)
    Y.append(y)
    
print(X)
print(Y)
```

```python
# Creating the scatter plot using plt.scatter()

plt.scatter(X,Y, marker='x', color='red', s=50) #s is the size of the marker
plt.show()
```

```python
# Now, creating a scatter plot with many values
x = []
y = []

np.random.seed(0) # seed for reproducibility
for i in range(0,100):
    x.append(np.random.randint(1,100))
    y.append(np.random.randint(1,100))
    
plt.scatter(x, y, marker='o', s=50, color='g' , alpha=0.60)
plt.xlim(0,100)
plt.ylim(0,100)
plt.show()
```

##### NOTE: As the data gets larger, it is advised to use **`plt.plot()`** over **`plt.scatter`**.  The reason is that **`plt.scatter()`** has capability to render a different size and/or color for each point, so the renderer must do extra work of constructing each point individually. In **`plt.plot()`** on the other hand, the points are always essentially clones of one another. So the work in determining the appearance of points is done only once per entire set of data.

### Boxplots (also used to understand distribution and perform uni-variate analysis)

- **`Interquartile range (IQR):`** These data points range between the 25th and 75th percentile values.

### HeatMaps

- A heat map (or heatmap) is a data visualization technique that shows magnitude of a phenomenon as color in two dimensions. The variation in color may be by hue or intensity, giving obvious visual cues to the reader about how the phenomenon is clustered or varies over space.

##### Additional Notes:

- Heat maps are predominantly used in machine learning problems to visualise a  Correlation Matrix, a grid that shows the correlation between any two quantitative variables. As mentioned in the additional notes of previous segments, understanding the correlation between variables is crucial for building and evaluating any ML model.
